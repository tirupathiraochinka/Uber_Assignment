Problem Statement:
Given the dataset containing the rides information for 5 consecutive days observed at places like within the City and at Airport, we need to analyse on the different status avaialble in the dataset and have to find out the locationwise time periods at which Uber is lossing its businees due to Unavailablity of the cabs.

Different Features anvailable in the dataset:
request_id        : Once you select your pickup_point, drop_point and confirm your ride,you will be assigned with an Request Id                     with which your request will be taken. You can contact the Uber Customer Support with this ID in case of                       any issues furtherly. 
pickup_point	  : The location at which you want to board the cab service. 
driver_id         : Once request_id is generated and it will be routed to the nearby driver available in that location. If the                     driver accepts the request then the driver_id will be assigned to the particular request_id
status	          : Here we have three different status observed per each request, namely TripCompleted,No Cars Available,                         Cancelled. If the passenger is able to complete the ride to specified drop locaiton then that request_id                       will get the status Trip Completed.
                    If there were no drivers availbale at the specified location to pick the passenger, the request_id will be                     having the status: No Cars Available.
                    In few cases passenger may find the driver is taking more time to reach the location and cancel the ride.                       In other cases driver may also cancel the ride if he feels he is earning less by taking this ride.  
request_timestamp : The time at which the passenger requests for a ride.This will have date and time information.
drop_timestamp    : The time at which the passenger completes the ride. This will be null for the rides which were not                             successful.

# Supply Demand Gap Analysis

Before making analysis on Supply Demand Gap, first let us understand which of the following status in our dataset comes under Demand and which status comes under Supply.
<p>Trip Completed:This states that the rider has requested for a ride and Uber is successfull in making the rider's request complete.In this case Uber was able to meet the demand of the user.
<p>No Cars Available: This is when the rider requests for the ride and when Uber has no cars available to meet the rider requests it will gives the message like 'No Cars Available'.In this case Demand is exceeding Supply which leading to gap between Supply and Demand. This needs the attention of the business Necessary Steps need to be taken to reduce this gap as Uber is loosing its customers due to lack of analysis.
<p>Cancelled:This will maily occur in two cases.First case is when Rider makes a request for a cab and when he/she feels the ETA(Estimated Time of Arrival) is high or when the rider has found any other means of transaport better to this  to cancel the ride and changes his/her mood.Second Case is when Driver is not willing to take the ride when he has any personal or geographical issues, he may cancel the ride at his end or asks the rider to cancel the ride. In first case, cabs are available but are at longer distances leading to Cancellation of the ride. In second case though the Cabs are available and demand was met, due to issues with driver, requests are getting Cancelled.As we are not aware due to which reason the ride was cancelled, it is better to consider both the cases as failure of Uber in predicting the rides inorder not to loose to the business.



So creating a new feature 'Supply' to consists of only the 'Trip Completed' data  

```python
impu_uber_pivot['supply']=impu_uber_pivot['Trip Completed']
```

Feature 'demand' to consists of all the data like 'Trip Completed','No Cars Available' and 'Cancelled'

```python
impu_uber_pivot['demand']=impu_uber_pivot['Trip Completed']+impu_uber_pivot['No Cars Available']+ impu_uber_pivot['Cancelled']
```

Feature 'supply_demand_gap' to consists of the difference between supply and demand.


```python
impu_uber_pivot['supply_demand_gap']=impu_uber_pivot['supply']-impu_uber_pivot['demand']
```

```python
impu_uber_pivot.head()
```

Negative Values in 'supply_demand_gap' features states that Demand is exceeding the Supply.

```python
for i in impu_uber_pivot.index.levels[0]:
    impu_uber_pivot.loc[i,:].groupby(level=0).plot(kind='bar',figsize=(15,10),title='Analysis including Traffic,Supply Demand Gap at '+ i)
for ax in plt.gcf().axes:
    plt.sca(ax)
    plt.xticks(rotation=45)
```
